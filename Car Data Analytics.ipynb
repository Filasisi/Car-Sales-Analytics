## Get car Manufacturer from "CarName" column

```python
df['CarName'] = df['CarName'].apply(lambda x: x.split(' ')[0])
```

```python
df.head()
```

### Rename CarName to Manufacturer

```python
df = df.rename(columns={'CarName': 'manufacturer'})
```

```python
df.head()
```

```python
df['manufacturer'].unique()
```

### Replace Manufacturer's name that are wrongly spelt



```python
#This code is used to correct spelling mistakes for manufacturers name 

def replace_name(a,b):
    df['manufacturer'].replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('Nissan','nissan')
replace_name('toyouta','toyota')
replace_name('porcshce','porsche')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')
```

```python
# check to ensure all spoking mistakes have been corrected
df['manufacturer'].unique()
```

## Frequency of Car Sold

```python

plt.figure(figsize=(30, 8))
plt1=sns.countplot(x=df.manufacturer, data=df, order= df.manufacturer.value_counts().index)
plt.title('Number of Car Sold by Manufacturer', size=25)
plt1.set_xlabel('Manufacturer', fontsize=19)
plt1.set_ylabel('Number of Car Sold', fontsize=19)
plt1.set_xticklabels(plt1.get_xticklabels(),rotation=360, size=14)
plt.show()
```

```python
def box_plot(data, x_column, y_column = "price"):
    sns.boxplot(x = x_column, y = y_column, data = df)
    plt.title(str(x_column).title() + " and " + str(y_column).title(), fontsize = 22)
    plt.xlabel(x_column.title(), fontsize = 15)
    plt.ylabel(y_column.title(), fontsize = 15)
```

```python
plt.figure(figsize = (10, 10))
box_plot(df, x_column = "price", y_column = "manufacturer")
```

##### The boxplot shows that BMW, Jaguar, Buick and Porshe seem to be more pricy that the other cars sold by Oxmoor Car Dealership in 2020. Also, it shows that the most expensive car that was sold by OCD was maunfactured by Buick.
